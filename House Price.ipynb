```python
train_df=pd.read_csv("train.csv")
test_df=pd.read_csv("test.csv")
sample_op=pd.read_csv("sample_submission.csv")
```

```python
train_df.LotFrontage.fillna(value=int(train_df.LotFrontage.mean()),inplace=True)
train_df.MasVnrArea.fillna(value=0.0,inplace=True)
train_df.Electrical.fillna(value='SBrkr',inplace=True)
```

```python
train_df.info()
```

```python
train_df.GarageCond.fillna(value="TA",inplace=True)
train_df.GarageQual.fillna(value="TA",inplace=True)
train_df.GarageType.fillna(value="Attchd",inplace=True)
train_df.GarageYrBlt.fillna(value=1978,inplace=True)
train_df.GarageFinish.fillna(value="Unf",inplace=True)
```
